<h2>--- Day 12: The N-Body Problem ---</h2>
<p>The space near Jupiter is not a very safe place; you need to be careful of a <a href="https://en.wikipedia.org/wiki/Great_Red_Spot">big distracting red spot</a>, extreme <a href="https://en.wikipedia.org/wiki/Magnetosphere_of_Jupiter">radiation</a>, and
    a <a href="https://en.wikipedia.org/wiki/Moons_of_Jupiter#List">whole lot of moons</a> swirling around. You decide to start by tracking the four largest moons: <em>Io</em>, <em>Europa</em>, <em>Ganymede</em>, and <em>Callisto</em>.</p>
<p>After a brief scan, you calculate the <em>position of each moon</em> (your puzzle input). You just need to <em>simulate their motion</em> so you can <span title="Or you could just leave, but where's the fun in that?">avoid them</span>.</p>
<p>Each moon has a 3-dimensional position (<code>x</code>, <code>y</code>, and <code>z</code>) and a 3-dimensional velocity. The position of each moon is given in your scan; the <code>x</code>, <code>y</code>, and <code>z</code> velocity of each moon starts
    at <code>0</code>.</p>
<p>Simulate the motion of the moons in <em>time steps</em>. Within each time step, first update the velocity of every moon by applying <em>gravity</em>. Then, once all moons' velocities have been updated, update the position of every moon by applying <em>velocity</em>.
    Time progresses by one step once all of the positions are updated.</p>
<p>To apply <em>gravity</em>, consider every <em>pair</em> of moons. On each axis (<code>x</code>, <code>y</code>, and <code>z</code>), the velocity of each moon changes by <em>exactly +1 or -1</em> to pull the moons together. For example, if Ganymede has
    an <code>x</code> position of <code>3</code>, and Callisto has a <code>x</code> position of <code>5</code>, then Ganymede's <code>x</code> velocity <em>changes by +1</em> (because <code>5 &gt; 3</code>) and Callisto's <code>x</code> velocity <em>changes by -1</em>    (because <code>3 &lt; 5</code>). However, if the positions on a given axis are the same, the velocity on that axis <em>does not change</em> for that pair of moons.</p>
<p>Once all gravity has been applied, apply <em>velocity</em>: simply add the velocity of each moon to its own position. For example, if Europa has a position of <code>x=1, y=2, z=3</code> and a velocity of <code>x=-2, y=0,z=3</code>, then its new position
    would be <code>x=-1, y=2, z=6</code>. This process does not modify the velocity of any moon.</p>

```python
import os
import re
import unittest

class ThreeDimensions():
    """ Data structure with 3 dimensions. """
    def __init__(self, x, y, z):
        self.x = int(x)
        self.y = int(y)
        self.z = int(z)
        
    def __str__(self):
        return f'(x:{self.x}, y:{self.y}, z:{self.z})'

    def __repr__(self):
        return self.__str__()

    def __eq__(self, other: 'ThreeDimensions'):
        """ Is equal operator """
        return (self.x, self.y, self.z) == (other.x, other.y, other.z)

    def __iadd__(self, other: 'ThreeDimensions'):
        """ addition assignment operator """
        self.x += other.x
        self.y += other.y
        self.z += other.z
        return self
    
    def __iter__(self):
        """ initializtion for iteration """
        self.n = 0
        return self
    
    def __next__(self):
        """ iteration process """
        if self.n < 3:
            result = (self.x, self.y, self.z)[self.n]
            self.n += 1
            return result
        else:
            raise StopIteration

    def right(self):
        self.x += 1

    def left(self):
        self.x -= 1

    def up(self):
        self.y += 1

    def down(self):
        self.y -= 1

    def forward(self):
        self.z += 1

    def back(self):
        self.z -= 1    

class Body():
    def __init__(self, position, velocity, name = ""):
        self.position = position
        self.velocity = velocity
        self.name = name
        
    def __str__(self):
        return f'Body(position:{self.position}, velocity:{self.velocity})'

    def __repr__(self):
        return self.__str__()

    def __eq__(self, other: 'Body'):
        """ Is equal operator """
        return (self.position, self.velocity) == (other.position, other.velocity)
    
    def get_total_energy(self):
        """ Returns the total energy of itself """
        potential_energy = sum(abs(val) for val in self.position)
        kinetic_energy = sum(abs(val) for val in self.velocity)
        return potential_energy * kinetic_energy
    
    def diff(self, other):
        """ Returns a diff for another body """
        
        x_diff = (self.position.x == other.position.x and self.velocity.x == other.velocity.x)
        y_diff = (self.position.y == other.position.y and self.velocity.y == other.velocity.y)
        z_diff = (self.position.z == other.position.z and self.velocity.z == other.velocity.z)
        
        return (x_diff,y_diff,z_diff)
        
    def update_velocity(self, other: 'ThreeDimensions'):
        """ Updates the velocity based on the simple gravity model """
        
        if self == other:
            return
        
        if self.position.x < other.position.x:
            self.velocity.right()
            other.velocity.left()
        elif self.position.x > other.position.x:
            self.velocity.left()
            other.velocity.right()

        if self.position.y < other.position.y:
            self.velocity.up()
            other.velocity.down()
        elif self.position.y > other.position.y:
            self.velocity.down()
            other.velocity.up()

        if self.position.z < other.position.z:
            self.velocity.forward()
            other.velocity.back()
        elif self.position.z > other.position.z:
            self.velocity.back()
            other.velocity.forward()
            
    def update_position(self):
        """ Updates the position based the simple velocity model """
        self.position += self.velocity

class NbodyOrbitSimulator():
    
    def __init__(self, data = '', filename = ''):
        self.bodies = []
        if data:
            self.load_bodies(data)
        elif filename:
            self.load_bodies_from_file(filename)
        else:
            pass
        
    def __str__(self):
        str = ''
        for body in self.bodies:
            str += f'{body}\n'
        return str

    def __repr__(self):
        return self.__str__()

    def load_bodies_from_file(self, filename):
        with open(filename) as file:
            data = file.read()            
        self.load_bodies(data)
    
    def load_bodies(self, data):
        """ Parses a string where each new line contains the 3 dimentional position and velocity of a body. """
        re_prog = re.compile('<x=(-?\d+), y=(-?\d+), z=(-?\d+)>')
        for x, y, z in re_prog.findall(data):
            self.bodies.append(Body(position=ThreeDimensions(x,y,z),velocity=ThreeDimensions(0,0,0)))
        
    def update_velocities(self):
        """ Updates the velocity of each body """
        for i1, body1 in enumerate(self.bodies):
            for body2 in self.bodies[i1+1:]:
                body1.update_velocity(body2)
                
    def update_positions(self):
        """ Updates the velocity of each body """
        for body in self.bodies:
            body.update_position()
        
    def step(self):
        """ Performs a single simulation step by updating the velocities and positions of each body """
        self.update_velocities()
        self.update_positions()
        
    def run(self, steps = 1, verbose=False):
        """ Runs the simulation for the given number of steps """
        if verbose:
            print(self)
            
        for _ in range(steps):
            self.step()
            
            if verbose:
                print(self)
            
    def get_total_energy(self):
        """ Returns the total energy of all bodies """
        total_energy = 0
        for body in self.bodies:
            total_energy += body.get_total_energy()

        return total_energy    
    

    def diff_bodies(self, other_bodies):
        """ Returns a list of booleans for the difference in dimentions """
        
        if len(self.bodies) != len(other_bodies):
            raise Exception('Cannot diff unequal bodies sizes')
        
        diffs = [self.bodies[i].diff(other_bodies[i]) for i in range(len(self.bodies))]
        
        return [all(vals) for vals in zip(*diffs)]
            
        

class TestSim(unittest.TestCase):
    def test_total_energy(self):
        test_data = [("<x=-1, y=0, z=2><x=2, y=-10, z=-7><x=4, y=-8, z=8><x=3, y=5, z=-1>",
                      [(0, 0), (1, 229), (2, 245), (3, 202), (4, 169), (5, 191), (6, 150), (7, 289), (8, 275), (9, 142), (10, 179)]),
                     ("<x=-8, y=-10, z=0><x=5, y=5, z=10><x=2, y=-7, z=3><x=9, y=-8, z=-3>",
                      [(0, 0), (10, 706), (20, 502), (30, 732), (40, 953), (50, 1292), (60, 1397), (70, 1634), (80, 939), (90, 1435), (100, 1940)])]
        for bodies_data, data in test_data:
            for sim_steps, result in data:
                sim = NbodyOrbitSimulator(data=bodies_data)
                sim.run(steps=sim_steps, verbose=False)
                self.assertEqual(sim.get_total_energy(), result)
            
unittest.main(argv=[""], exit=False)
```

<p><em>How many steps does it take</em> to reach the first state that exactly matches a previous state?</p>

```python
import numpy as np
import copy

sim = NbodyOrbitSimulator(filename='inputs/input_d12.txt')
initial_bodies = copy.deepcopy(sim.bodies)
match_steps_x = None
match_steps_y = None
match_steps_z = None

step = 0
while not all((match_steps_x, match_steps_y, match_steps_z)):
    sim.run()
    step += 1
    
    x_diff, y_diff, z_diff = sim.diff_bodies(initial_bodies)

    if x_diff and match_steps_x is None:
        match_steps_x = step

    if y_diff and match_steps_y is None:
        match_steps_y = step

    if z_diff and match_steps_z is None:
        match_steps_z = step

steps_till_initial = np.lcm(np.lcm(match_steps_x, match_steps_y), match_steps_z)
print(f'Simulation ran for {step} steps')
print(f'Number of steps till initial: {steps_till_initial}')
```
